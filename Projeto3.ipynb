```python
print('Bem vindo a Livraria do João Vitor Potratz Pereira')

lista_livro = []
id_global = 0


def cadastrar_livro(id):
# Cadastra um novo livro na lista_livro.

    print('----------------------------------------------------------')
    print('--------------- MENU CADASTRAR LIVRO ---------------------')
    print(f'Id do livro: {id}')

    nome = input('Por favor digite o nome do livro: ')
    autor = input('Por favor digite o autor do livro: ')
    editora = input('Por favor digite a editora do livro: ')
    print('----------------------------------------------------------\n')

    livro = {'id': id, 'nome': nome, 'autor': autor, 'editora': editora}

    lista_livro.append(livro.copy())


def consultar_livro():
# Consulta livros por critério solicitado.

    while True:
        print('----------------------------------------------------------')
        print('--------------- MENU CONSULTAR LIVRO ---------------------')
        opcao = input(f'1 - Consultar Todos os Livros\n'
                      f'2 - Consultar Livro por id\n'
                      f'3 - Consultar Livro(s) por autor\n'
                      f'4 - Retornar\n'
                      f'>>')

        if opcao == '1':
            if not lista_livro:
                print('----------------------------------------------------------')
                print('Não há livros cadastrados.\n')
            else:
                print('----------------------------------------------------------')
                for livro in lista_livro:
                   print(f"id: {livro['id']}\n"
                          f"nome: {livro['nome']}\n"
                          f"autor: {livro['autor']}\n"
                          f"editora: {livro['editora']}\n")
                print('----------------------------------------------------------')
        elif opcao == '2':
            try:
                id_consulta = int(input('Digite o id do livro: '))
                print('----------------------------------------------------------')
                encontrado = False
                for livro in lista_livro:
                    if livro['id'] == id_consulta:
                        print(f"id: {livro['id']}\n"
                              f"nome: {livro['nome']}\n"
                              f"autor: {livro['autor']}\n"
                              f"editora: {livro['editora']}\n")
                        print('----------------------------------------------------------')
                        encontrado = True
                        break
                if not encontrado:
                    print('Nenhum livro encontrado com o id informado.\n')
                    print('----------------------------------------------------------')
            except ValueError:
                print('----------------------------------------------------------')
                print('Por favor, digite um id válido (número inteiro).\n')
                print('----------------------------------------------------------')
        elif opcao == '3':
            autor_consulta = input('Digite o autor do(s) livro(s): ')
            print('----------------------------------------------------------')
            livros_autor = [livro for livro in lista_livro
                            if livro['autor'].lower() == autor_consulta.lower()]
            if livros_autor:
                for livro in livros_autor:
                    print(f"id: {livro['id']}\n"
                          f"nome: {livro['nome']}\n"
                          f"autor: {livro['autor']}\n"
                          f"editora: {livro['editora']}\n")
                print('----------------------------------------------------------')
            else:
                print('----------------------------------------------------------')
                print(f'Nenhum livro encontrado do autor "{autor_consulta}".\n')
                print('----------------------------------------------------------')
        elif opcao == '4':
            # retorna ao menu principal
            break
        else:
            # Vai repetir a pergunta
            print('Opção inválida!')


def remover_livro():
# Permite o usuário remover um livro da lista_livro pelo ID

    while True:
        print('----------------------------------------------------------')
        print('---------------- MENU REMOVER LIVRO ----------------------')
        id_remover = input('Digite o id do livro a ser removido: ')
        try:
            id_remover = int(id_remover)
            indice_remover = -1
            for i, livro in enumerate(lista_livro):
                if livro['id'] == id_remover:
                    indice_remover = i
                    break
            if indice_remover != -1:
                lista_livro.pop(indice_remover)
                print('Livro removido com sucesso!\n')
                break
            else:
                # Vai repetir a pergunta
                print("Id inválido. Nenhum livro encontrado com o ID informado.\n")
        except ValueError:
            print('Por favor, digite um ID válido (número inteiro)\n')


while True:
    print('----------------------------------------------------------')
    print('------------------- MENU PRINCIPAL -----------------------')
    opcao = input(f'Escolha a opção desejada:\n'
                  f'1 - Cadastrar Livro\n'
                  f'2 - Consultar Livro(s)\n'
                  f'3 - Remover Livro\n'
                  f'4 - Sair\n'
                  f'>>')

    if opcao == '1':
        # incrementa o ID global antes de cadastrar um novo livro
        id_global += 1
        cadastrar_livro(id_global)
    elif opcao == '2':
        consultar_livro()
    elif opcao == '3':
        remover_livro()
    elif opcao == '4':
        print('Obrigado e volte sempre!')
        # sai do loop do menu principal e encerra o programa
        break
    else:
        # vai repetir a pergunta
        print('Opção inválida!\n')
```
